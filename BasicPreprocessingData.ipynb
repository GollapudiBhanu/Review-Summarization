```python
#You can write you answer here. (No code needed)
import pandas as pd

blinkMini_review = pd.read_csv("BlinkMini.csv")
blinkMini_review = blinkMini_review.dropna()

print(blinkMini_review.shape)
print(blinkMini_review.head())
```

```python
from nltk.corpus import stopwords
from textblob import TextBlob
from nltk.stem import PorterStemmer
from textblob import Word
import re

nltk.download('wordnet')
st = PorterStemmer()
nltk.download('punkt')
stop = stopwords.words('english')

def clean_data(df, column):
  # Lower case
  print(df)
  print(column)
  df[column] = df[column].apply(lambda x: " ".join(x.lower() for x in str(x).split()))
  #Removing Punctuation
  df[column] = df[column].str.replace('[^\w\s]','')
  #Removing special charachters
  df[column] = df[column].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
  #Removing stopwords
  df[column] = df[column].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
  #Tokenization
  df[column] = df[column].apply(lambda x: TextBlob(x).words)
  #Lemmatization
  df[column] = df[column].apply(lambda x: " ".join([Word(word).lemmatize() for word in x]))
  return df
```

```python
br_columns = ["Review Title", "Review Text"]
t_blinkMini_review = blinkMini_review
for c in br_columns:
  blinkMini_review = clean_data(blinkMini_review, c)

blinkMini_review.head()
```

```python
#blinkMini_review.to_csv(index=False)
blinkMini_review.to_csv('BlinkMini_clean.csv', index=False)
```
