```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import wordcloud
```

```python
df = pd.read_csv("SecurityDevices_Full.csv")
df.head()
```

## BlinkMini

```python
data_blinkmini = df[(df.Device_Name == 'BlinkMini')]
data_blinkmini_list = data_blinkmini["Total_Description"].values.tolist()
blinkmini_reviews = ' '.join([str(item) for item in data_blinkmini_list])
word_cloud_blinkmini = WordCloud(width = 2000, height = 1000, collocations = False, background_color = 'white').generate(blinkmini_reviews)
plt.figure(figsize = (20,10))
plt.imshow(word_cloud_blinkmini, interpolation='bilinear')
plt.axis("off")
plt.show()
```
